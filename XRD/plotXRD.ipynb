#  XRD chart


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# crystallographic cards info from Ag and Cu
Ag = pd.read_csv('Ag.txt', header=None, sep='\t')
Cu = pd.read_csv('Cu.txt', header=None, sep='\t')
```

```python
# read of the asc file
df1 = np.loadtxt('CuAg_Xrd.asc', skiprows=2)
```

```python
#creation of dataframe
df = pd.DataFrame(df1)
```

```python
#to create several charts in one figure
fig, ax = plt.subplots(3, sharex=True, figsize=(14,12),  gridspec_kw={'height_ratios': [3, 1,1]})  
plt.rc('legend', fontsize='small') # legend fontsize

# First subfigure
ax[0].plot(df[0], df[1], label="CuAg")
ax[0].set_ylim(-1,3500)  #set the limit of the y axis
ax[0].legend(loc="best",fontsize=18)
ax[0].minorticks_on()
ax[0].tick_params(axis='both', direction='out', which='both',right=True,top=True)
ax[0].set_ylabel('a.u',fontsize=12)

# Point identification
ax[0].text(37.117,3000, 'Ag(111)')
ax[0].text(43.342,1100, 'Ag(200)')
ax[0].text(64.428,950, 'Ag(220)')
ax[0].text(77.475,950, 'Ag(311)')
ax[0].text(81.539,400, 'Ag(222)')
ax[0].text(95.891,300, 'Ag(400)')
ax[0].grid(True)


#Subplot for Ag
# Ag
ax[1].bar(Ag[0], Ag[1],width=0.1, label="Ag")
ax[1].legend(loc="best",fontsize=14)
ax[1].tick_params(axis='both', direction='out', which='both',right=True,top=True)
ax[1].set_ylabel('%',fontsize=12)

# Subplot for Cu
ax[2].bar(Cu[0],Cu[1],width=0.1,label="Cu")
ax[2].legend(loc="best",fontsize=14)
ax[2].tick_params(axis='both', direction='out', which='both',right=True,top=True)
ax[2].set_ylabel('%',fontsize=12)
ax[2].set_xlabel("2$\Theta$(°)",fontsize=20)


fig.suptitle('Difracción de rayos X del material CuAg', fontsize=27)
fig.supylabel("Intensidad",fontsize=20)

plt.xlim(30,100)
plt.savefig("CuAg_XRD.png", dpi=300)
plt.show()
```
